```python
low = 137683
hig = 596253

count_valid = 0
for pw in range(low, hig + 1):
    is_valid = True
    has_double = False
    prev_c = -1

    for c in map(int, str(pw)):
        if c < prev_c: 
            is_valid = False
            
        if c == prev_c:
            has_double = True
                
        prev_c = c
        
    if not has_double: 
        is_valid = False
        
    if is_valid:
        count_valid += 1
        # print(pw)
        
print(f'Answer 1: {count_valid}')
```

```python
count_valid = 0
for pw in range(low, hig + 1):
    is_valid = True
    prev_c = -1
    found_doubles = []
    double_count = 0

    for c in map(int, str(pw)):
        if c < prev_c: 
            is_valid = False
        
        if c == prev_c:
            double_count += 1
            if double_count == 1:
                found_doubles.append(c)
            elif double_count > 1 and c in found_doubles:
                found_doubles.remove(c)
        else:
            double_count = 0
                
        prev_c = c
        
    if len(found_doubles) == 0: 
        is_valid = False
        
    if is_valid:
        count_valid += 1
        print(pw)
        
print(f'Answer 2: {count_valid}')
```
